```python
# Define the categorical variables of the df to display in pie charts

categorical_variables = ['gender', 'hypertension', 'heart_disease', 'ever_married', 'work_type', 'Residence_type', 'smoking_status', 'stroke']

# Create subplots for each variable
num_rows = 2  # Number of rows of subplots
num_cols = 4  # Number of columns of subplots
fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Flatten the subplots axes
axs = axs.flatten()

# Iterate over the variables and create pie charts
for i, var in enumerate(categorical_variables):
    counts = df[var].value_counts()
    labels = counts.index.tolist()
    sizes = counts.values.tolist()
    axs[i].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    axs[i].set_title(var)
    
# Adjust the layout and spacing
plt.tight_layout()
```

```python
df = pd.read_csv('healthcare-dataset-stroke-data.csv')
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

df
```

### Label Encoding 
Label Encoding is a technique that is used to convert categorical columns into numerical ones so that they can be fitted by machine learning models which only take numerical data. It is an important pre-processing step in a machine-learning project.
For example; a column having elements as low, medium, or high could be replaced with 1,2,3, where 1 represents 'low', 2 'medium', and 3 'high'. 

```python
print(df.columns.tolist())
```

```python
# Turn categorical data into numerical so we can run a model
# from sklearn.preprocessing import LabelEncoder

from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in each column 
df['gender']= label_encoder.fit_transform(df['gender'])
df['ever_married']= label_encoder.fit_transform(df['ever_married'])
df['work_type']= label_encoder.fit_transform(df['work_type'])
df['Residence_type']= label_encoder.fit_transform(df['Residence_type'])
df['smoking_status']= label_encoder.fit_transform(df['smoking_status'])
```

```python
df
```

### Data spilting and training model

```python
# Define your target data and variables
x = df.drop('stroke', axis=1)     # This takes all data but the target data
y = df['stroke']                  # This takes just the target data
```

```python
# Split the data into test and train
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
# SMOTE (Synthetic Minority Over-sampling Technique) is a popular technique used to address class imbalance in ML tasks, 
# particularly when the minority class has limited representation.
# !pip install imbalanced-learn

from imblearn.over_sampling import SMOTE

smt = SMOTE(random_state = 0)
x_train_res, y_train_res = smt.fit_resample(x_train, y_train)
```

### Run a Regression model

How to choose the right model: https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html

```python
# Run a logistic regression model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

model = LogisticRegression()

model.fit(x_train_res, y_train_res)

print(model.score(x_train_res, y_train_res))
pred = model.predict(x_test)
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier()

model.fit(x_train,y_train)

print(model.score(x_test,y_test))
pred1=model.predict(x_test)
```

### Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()

model.fit(x_train,y_train)

print(model.score(x_test,y_test))

pred2=model.predict(x_test)
```

### OLS model

```python
#X = sm.add_constant(x)
import statsmodels.api as sm
X = sm.add_constant(x)

model = sm.OLS(y, X)
model_result = model.fit()
model_result.summary()
```

In this data, the Random forest classifier gave us the best accuracy (96%) 

> This means that the model can redict if a patent is at risk of a stroke with an accuracy of 96%!
